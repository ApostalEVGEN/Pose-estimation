# Детекция позы сноубордиста

#### Загрузка библиотек

```python
from ultralytics import YOLO
import cv2
import torch
from IPython.display import display, Image  # Для отображения изображений в Jupiter
```

```python
# pip freeze > requirements.txt сохраним версии библиотек используемые на данный момент
```

#### Загрузка предобученной модели YOLO

```python
# Загрузите обученную модель
model = YOLO('yolov8n-pose.pt')  # Обновите путь к вашей модели
```

```python
# Переместите модель на нужное устройство (используйте "cpu", если CUDA недоступна).
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model.to(device)

# Определите путь к видео
video_path = 'Snowboard.mp4'  # Замените на путь к вашему видеофайлу

# Откройте видео с помощью OpenCV
cap = cv2.VideoCapture(video_path)

# Получение свойств видео (ширина кадра, высота и частота кадров в секунду)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Определите кодек и создайте объект VideoWriter для сохранения выходных данных
# output_path = f'{video_path}output_video_pose.mp4'  # Определите свой выходной путь
output_path = f'{video_path}Snowboard_pose.mp4'  # Определите свой выходной путь
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Кодек
out = cv2.VideoWriter(output_path, fourcc, fps, (width // 2, height // 2))  # Сохранить в половинном размере
```

#### Обработка видео

```python
# Прокрутите каждый кадр в видео
frame_count = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Прервите цикл, если больше нет кадров

    # Уменьшите размер рамки вдвое, чтобы уменьшить потребление памяти
    frame = cv2.resize(frame, (width // 2, height // 2))

    # Проведите кадром по модели для определения позы
    results = model(frame)

    # Извлеките и визуализируйте позы на кадре
    annotated_frame = results[0].plot()  # Нанесение ключевых точек и каркасов на кадр

    # Запишите аннотированный кадр в выходное видео
    out.write(annotated_frame)

    # Отображать каждый 50-й кадр в Jupiter (необязательно, закомментируйте это, чтобы избежать отображения)
    if frame_count % 50 == 0:  # Отрегулируйте частоту отображения кадров по мере необходимости
        _, encoded_img = cv2.imencode('.jpg', annotated_frame)  # Закодировать в формате jpg
        display(Image(data=encoded_img.tobytes()))  #Рамка дисплея в Jupyter

    # Освобождайте память графического процессора после каждого кадра, чтобы предотвратить перегрузку памяти
    torch.cuda.empty_cache()

    frame_count += 1

# Release video resources
cap.release()
out.release()

print("Обработка видео завершена. Выходные данные сохранить как", output_path)
```
